```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta
import os

fake = Faker()

# Crear carpeta data si no existe
os.makedirs("../data", exist_ok=True)

# Parámetros
num_rows = 2000

products = [
    ("Laptop Pro 15", "Electronics"),
    ("Laptop Air 13", "Electronics"),
    ("Wireless Mouse", "Electronics"),
    ("Office Chair", "Furniture"),
    ("Standing Desk", "Furniture"),
    ("Water Bottle", "Accessories"),
    ("Noise Cancelling Headphones", "Electronics"),
    ("Backpack Travel", "Accessories"),
    ("Coffee Maker", "Home"),
    ("Smartwatch", "Electronics"),
]

payment_methods = ["Credit Card", "Debit Card", "PayPal", "Apple Pay", "Cash"]
cities = ["Toronto", "Vancouver", "Mexico City", "New York", "Chicago", "Shanghai"]

data = []

start_date = datetime(2023, 1, 1)

for i in range(num_rows):
    product, category = random.choice(products)
    price = round(random.uniform(10, 2500), 2)
    quantity = random.randint(1, 4)
    date = start_date + timedelta(days=random.randint(0, 700))

    row = {
        "order_id": fake.uuid4(),
        "order_date": date.strftime("%Y-%m-%d"),
        "product": product,
        "category": category,
        "price": price,
        "quantity": quantity,
        "customer_id": fake.random_int(min=1000, max=9999),
        "city": random.choice(cities),
        "payment_method": random.choice(payment_methods),
    }

    data.append(row)

df = pd.DataFrame(data)
df.to_csv("../data/sales_data.csv", index=False)

df.head()
```

```python
import pandas as pd

df = pd.read_csv("../data/sales_data.csv")
df.head()
```

```python
# Ver cuántos valores faltantes hay
df.isna().sum()
```

```python
df = df.dropna()  # elimina filas con NaN
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
df["year_month"] = df["order_date"].dt.to_period("M")
```

```python
df["revenue"] = df["price"] * df["quantity"]
df["revenue"].sum()
```

```python
sales_by_month = df.groupby("year_month")["revenue"].sum()
sales_by_month.head()
#Ventas por mes
```

```python
top_products = df.groupby("product")["quantity"].sum().sort_values(ascending=False).head(10)
top_products
#Top 10 productos mas vendidos 
```

```python
category_revenue = df.groupby("category")["revenue"].sum().sort_values(ascending=False)
category_revenue
#Categoria con mas revenue 
```

```python
avg_sales_per_customer = df.groupby("customer_id")["revenue"].sum().mean()
avg_sales_per_customer
#Promedio gastado por el cliente 
```

```python
sales_by_city = df.groupby("city")["revenue"].sum().sort_values(ascending=False)
sales_by_city
#Ventas por ciudad 
```

```python
payment_counts = df["payment_method"].value_counts()
payment_counts
#Metodos de pago mas usados 
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
#Librerias 
```

```python
import os

os.makedirs("../images", exist_ok=True)

```

```python
plt.figure(figsize=(12,6))
sales_by_month.plot(kind="line")
plt.title("Revenue por Mes")
plt.xlabel("Mes")
plt.ylabel("Revenue ($)")
plt.savefig("../images/revenue_por_mes.png", dpi=300, bbox_inches="tight")
plt.show()

#Grafico Ventas por mes 
```

```python
plt.figure(figsize=(12,6))
top_products.plot(kind="bar")
plt.title("Top 10 Productos Más Vendidos")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.show()
#Grafico Top productos
```

```python
plt.figure(figsize=(10,5))
category_revenue.plot(kind="bar", color="orange")
plt.title("Revenue por Categoría")
plt.xlabel("Categoría")
plt.ylabel("Revenue")
plt.show()
#Grafico Revenue por categoria 
```

```python
plt.figure(figsize=(10,5))
sales_by_city.plot(kind="bar", color="green")
plt.title("Revenue por Ciudad")
plt.xlabel("Ciudad")
plt.ylabel("Revenue")
plt.show()
#Grafico Ventas por ciudad
```
